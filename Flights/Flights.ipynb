```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Flight Data.csv')
df.head()
```

```python
#Check value if null
df.isnull().sum()
```

```python
#check duplicate value
df.duplicated
#remove duplicate if available
df.drop_duplicates()

df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.relplot(x='year', y='passengers', data=df, kind='line', hue='month')
plt.title('Filghts data year wise')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.barplot(x='month', y='passengers', data=df)
plt.title('Filghts data year wise')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.barplot(x='year', y='passengers', data=df)
plt.title('Filghts data year wise')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.barplot(x='month', y='passengers', data=df[:-1])
plt.title('Filghts data month wise of 1960')
plt.show()
```
