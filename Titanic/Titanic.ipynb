```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url = 'Titanic.csv'
df = pd.read_csv(url)
df.head(7)
```

```python
df.isnull().sum()
```

```python
dfs = df['age']
np.where(pd.isnull(dfs))
```

```python
df['age'].fillna(method='ffill',  inplace=True)
```

```python
df.isnull().sum()
```

```python
df['embarked'].ffill(axis=0, inplace=True)
df['embark_town'].bfill(axis=0, inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
new_df = df.rename(columns={'survived':'Survived', 'sex':'Sex', 'age':'Age', 'fare':'Fare'})
new_df
#But we will work with old Df
```

```python
# df.dropna(axis=1,  how='any')
df.drop(columns='deck')
```

#Data Visualization

```python
#all over data set
sns.barplot(data=df)
```

```python
sns.relplot(x='survived', y='age', hue='sex', data=df, kind='line')
plt.title('Survived people')
plt.show()
```

```python
sns.lineplot(x='age', y='fare', hue='class', data=df)
plt.title('Fair and class, Age wise')
plt.show()
```

```python
plt.subplots(figsize=(10,6))
sns.barplot(x='embark_town', y='age', hue='sex', data=df)
plt.title('Townwise travelling People')
plt.show()
```

```python
plt.subplots(figsize=(8,6))
sns.lineplot(x='class', y='survived', hue='sex', data=df, marker='*', markersize=20)
plt.title('people survived by class')
plt.show()
```

```python
sns.pairplot(hue='class', data=df)
plt.show()
```
